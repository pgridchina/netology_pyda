# 5.С использованием t-test статистики проверить взаимосвязи между несколькими переменными.

```python
# посмотрим без нормирования
df1 = df['promotion_last_5years']
df2 = df['satisfaction_level']
stats.ttest_ind(df1, df2,equal_var=False)
```

```python
stats.ttest_ind(df3, df4)
```

# 10.Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.


1. Видим, что уровень удовлетворения (satisfaction_level) у уволившихся сотрудников ниже. Проверим нулевую гипотезу, что уровень удовлетворенности не влияет на увольнение.

```python
left_df=df[df["left"]==1]["satisfaction_level"]
avh_df=df["satisfaction_level"]
```

```python
len(avh_df), len(left_df)
```

```python
avh_df = avh_df.sample(3571)
left_df = left_df.sample(3571)
```

```python
sts.ttest_ind(avh_df, left_df)
```

pvalue стремится к 0, значит гипотезу о том, что уровень удовлетворенности не влияет, отвергаем. Т.к. у нас еще и отрицательная корреляция, делаем вывод, что чем меньше уровень удовлетворенности, тем больше сотрудников увольняется.
