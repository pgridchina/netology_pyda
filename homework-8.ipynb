### Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import choice
import scipy.stats as sts
from scipy.optimize import minimize
%matplotlib inline
```

```python


# зададим нормально распределенную случайную величину
norm_rv1 = sts.norm(loc=2, scale=0.5)
norm_rv2 = sts.norm(loc=4, scale=0.5)
norm_rv3 = sts.norm(loc=2, scale=1)
norm_rv4 = sts.norm(loc=4, scale=1)
norm_rv5 = sts.norm(loc=2, scale=2)
x = np.linspace(0,10,100)
pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)
plt.plot(x, pdf1,label='2-0.5')
plt.plot(x, pdf2, label='4-0,5')
plt.plot(x, pdf3, label='2-1')
plt.plot(x, pdf4,label='4-1')
plt.plot(x, pdf5, label='2-2')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python

rv1 = sts.expon(loc=2, scale=0.5)
rv2 = sts.expon(loc=4, scale=0.5)
rv3 = sts.expon(loc=2, scale=1)
rv4 = sts.expon(loc=4, scale=1)
rv5 = sts.expon(loc=2, scale=2)
x = np.linspace(0,10,100)

pdf1 = rv1.pdf(x)
pdf2 = rv2.pdf(x)
pdf3 = rv3.pdf(x)
pdf4 = rv4.pdf(x)
pdf5 = rv5.pdf(x)
plt.plot(x, pdf1,label='2-0.5')
plt.plot(x, pdf2, label='4-0,5')
plt.plot(x, pdf3, label='2-1')
plt.plot(x, pdf4,label='4-1')
plt.plot(x, pdf5, label='2-2')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python

rv1 = sts.t(0.01)
rv2 = sts.t(0.2)
rv3 = sts.t(1)
rv4 = sts.t(10)
rv5 = sts.t(50)
x = np.linspace(-10,10,100)

pdf1 = rv1.pdf(x)
pdf2 = rv2.pdf(x)
pdf3 = rv3.pdf(x)
pdf4 = rv4.pdf(x)
pdf5 = rv5.pdf(x)
plt.plot(x, pdf1,label='0.01')
plt.plot(x, pdf2, label='0.2')
plt.plot(x, pdf3, label='1')
plt.plot(x, pdf4,label='10')
plt.plot(x, pdf5, label='50')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
